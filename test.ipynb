```python
import torch
device = torch.device(
    "mps" if torch.backends.mps.is_available()
    else "cuda" if torch.cuda.is_available()
    else "cpu"
)
print(f"Using device: {device}")
```

```python
x = torch.randn(10, 10)
x = x.to(device)
print(x)
```
